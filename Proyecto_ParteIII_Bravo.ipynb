# Análisis de las características fisicoquímicas del agua para determinar su potabilidad:
**Abstract**

  Las características fisicoquímicas del agua juegan un papel crucial en la determinación de su potabilidad. Si bien todas las características recopiladas (PH, dureza, sólidos, cloraminas, sulfato, conductividad, carbono orgánico, trihalometanos y turbidez) son importantes, algunas tienen un impacto más significativo en la salud humana y la calidad del agua para consumo. Sin embargo, en general, podemos establecer una priorización aproximada de la siguiente manera:

1.  Turbidez: La turbidez indica la presencia de partículas suspendidas en el agua, como sedimentos, microorganismos o materia orgánica. El agua turbia puede ser desagradable a la vista y al gusto, y puede albergar patógenos que causan enfermedades.
2.  PH: El pH del agua potable debe estar entre 6.5 y 8.5. Un pH demasiado ácido o alcalino puede corroer las tuberías y afectar el sabor del agua, además de tener posibles efectos en la salud.
3.  Trihalometanos: Los THM se han clasificado como posibles carcinógenos humanos por la EPA. Esto significa que existe alguna evidencia de que los THM pueden causar cáncer en los seres humanos. También pueden causar otros problemas de salud, como problemas reproductivos y problemas de desarrollo en los niños.
4.  Sulfato: En sí mismos no son tóxicos en niveles moderados. Sin embargo, niveles elevados pueden dar un sabor amargo al agua y, en algunos casos, actuar como laxante.
5.  Solidos: Niveles elevados de sólidos totales disueltos pueden indicar la presencia de contaminantes y afectar el sabor, el olor y su potabilidad.
6.  Cloraminas: Las cloraminas son un compuesto químico compuesto por cloro y amoniaco que se utilizan como desinfectante en el agua potable. La exposición a largo plazo puede desarrollar asma u otras enfermedades respiratorias. Usualmente, se utiliza para reducir la formación de subproductos como el Trihalometanos.
7.  Dureza: La dureza excesiva del agua puede ocasionar problemas gastrointestinales, incrustaciones en tuberías y electrodomésticos, y afectar el sabor del agua.
8.  Conductividad: La conductividad es una medida de la capacidad del agua para conducir electricidad, y está relacionada con la presencia de iones disueltos. Si bien puede ser un indicador de la presencia de contaminantes, no es un parámetro tan específico como otros mencionados anteriormente.
9.  Carbono orgánico total (COT): El carbono orgánico puede provenir de diversas fuentes, incluyendo materia vegetal en descomposición o contaminación industrial. Si bien puede afectar el sabor y el olor del agua, no es un indicador directo de riesgos para la salud.

**Objetivo**

Poder determinar la potabilidad del agua según las características fisicoquímicas de esta.

**Contexto comercial**

El gobierno de Venezuela está buscando poder crear un modelo que pueda determinar si ciertos cuerpos de agua dulce son aptos para el consumo humano, ya que en el territorio nacional se obtienen sin ser explotados, se desea disponer más de este recurso tan necesario para el desarrollo humano, para esto se ha puesto a disposición muestras de algunos cuerpos de agua para llevar a cabo este modelo.

**Podemos plantearnos las siguientes hipótesis:**

1. El agua con un PH dentro del rango de 6.5 y 8.5 debería ser potable.

2. Para altos valores de turbidez, trihalometanos, sulfato, sólidos, cloraminas y dureza, el agua no debería ser potable.

3. Mientras más cloramina tenga el agua, menos trihalometanos debería tener esta. Sin embargo, si se tienen altos valores de cualquiera de estos dos compuestos, el agua no debería ser potable.




**Para iniciar, se importan los paquetes y dataset necesarios.**

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

url = 'https://raw.githubusercontent.com/Pancho28/calidad-del-agua/main/water_potability.csv'
df=pd.read_csv(url)
df
```

**Información acerca del dataset:**


1.   ph: pH del agua (de 0 a 14).
2.   Hardness: Capacidad del agua para precipitar el jabón en mg/L.
3.  Solids: Sólidos totales disueltos en ppm.
4.  Chloraminess: Cantidad de Cloraminas en ppm.
5.  Sulfate: Cantidad de Sulfatos disueltos en mg/L.
6.  Conductivity: Conductividad eléctrica del agua en μS/cm.
7.  Organic_carbon: Cantidad de carbono orgánico en ppm.
8.  Trihalomethanes: Cantidad de trihalometanos en μg/L.
9.  Turbidity: Medida de la propiedad de emisión de luz del agua en NTU (mg/L).
10. Potability: Indica si el agua es segura para el consumo humano. Potable 1 y No potable 0.

**Unidad de medida:**


*   ppm: partes por millón.
*   μg/L: microgramo por litro.
*   mg/L: miligramo por litro.
*   μS/cm: microsiemens por centímetro.

También para mejorar la lectura, se cambian los valores de potabilidad a 'Potable' y 'No potable', luego la columna se define como tipo 'category' y, por último, se agregan líneas en los gráficos.




```python
sns.set_style(style="darkgrid", rc={"grid.color": ".5"})
df.Potability = df.Potability.replace({1: "Potable", 0: "No potable"})
df.Potability = df.Potability.astype("category")
df.info()
```

# Valores perdidos

**Se buscan los valores faltantes por columna, se nota que solo hay valores faltantes en tres columnas ph, Sulfate y Trihalomethanes.**

```python
print(df.isna().sum())
```

**Se calcula el coeficiente de variación de las columnas faltantes, sabiendo que es menor al 30%, se podría decir que la media es representativa para los datos y los datos faltantes se podrían reemplazar por la media correspondiente.**

```python
coef_var_ph = df.ph.std() / df.ph.mean() *100
coef_var_sulfate = df.Sulfate.std() / df.Sulfate.mean() *100
coef_var_trihalomethanes = df.Trihalomethanes.std() / df.Trihalomethanes.mean() *100
print(f"""Coeficiente de variacion de la columna ph: {coef_var_ph.round(2)},
Coeficiente de variacion de la columna Sulfate: {coef_var_sulfate.round(2)}
Coeficiente de variacion de la columna Trihalomethanes: {coef_var_trihalomethanes.round(2)}""")
```

**Se creó un dataframe cambiando los valores faltantes por la media correspondiente.**

```python
df_sin_nan = df.copy()
df_sin_nan.ph.fillna(df_sin_nan.ph.mean(), inplace=True)
df_sin_nan.Sulfate.fillna(df_sin_nan.Sulfate.mean(), inplace=True)
df_sin_nan.Trihalomethanes.fillna(df_sin_nan.Trihalomethanes.mean(), inplace=True)
print(df_sin_nan.isna().sum())
```

**Se realizó una comparación gráfica con el dataframe original y el que no tiene valores faltantes. Donde podemos observar que los valores medios crecieron ampliamente.**

```python
f, axs = plt.subplots(3, 2, figsize=(10, 10), gridspec_kw={"height_ratios": (.50, .50, .50)})
sns.histplot(data=df, x="ph", hue="Potability",  ax=axs[0,0])
sns.histplot(data=df_sin_nan, x="ph", hue="Potability",  ax=axs[0,1])
sns.histplot(data=df, x="Sulfate", hue="Potability",  ax=axs[1,0])
sns.histplot(data=df_sin_nan, x="Sulfate", hue="Potability",  ax=axs[1,1])
sns.histplot(data=df, x="Trihalomethanes", hue="Potability",  ax=axs[2,0])
sns.histplot(data=df_sin_nan, x="Trihalomethanes", hue="Potability",  ax=axs[2,1])
f.tight_layout()
plt.show()
```

# Modelo

**Se importan las librerías a utilizar para crear el modelo, y se definen las variables "X", "Y" y se dividen estas para entrenamiento y prueba.**

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

X = df_sin_nan.copy()
X = X.drop(["Potability"], axis=1)
y = df_sin_nan.Potability
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

**Se aplican dos algoritmos de clasificación, para saber cuál es el más adecuado para este problema. Se utilizará K-Nearest-Neighbor y árbol de decisión.**

```python
modelo_k = KNeighborsClassifier(n_neighbors=5)
modelo_k.fit(X_train, y_train)

preds_train_k = modelo_k.predict(X_train)
y_pred_k = modelo_k.predict(X_test)

modelo_arbol = DecisionTreeClassifier(max_depth=10)
modelo_arbol.fit(X_train, y_train)

preds_train_arbol = modelo_arbol.predict(X_train)
y_pred_arbol = modelo_arbol.predict(X_test)

print(metrics.classification_report(y_test, y_pred_k))
print(metrics.classification_report(y_test, y_pred_arbol))
```

**Conclusiones**
****
**Comparando los dos modelos, se observa un comportamiento similar con una ligera ventaja en el modelo de árbol de decisiones. Sin embargo, ambos modelos presentan un mejor rendimiento al clasificar el agua 'No Potable' en comparación con 'Potable'. Esta diferencia podría atribuirse a la mayor cantidad de muestra de agua 'No Potable', lo que podría estar afectando el rendimiento de los modelos.**
